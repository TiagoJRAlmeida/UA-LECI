```python
#import necessary modules
import pandas as pd
import json
from difflib import get_close_matches
from Levenshtein import distance
import numpy as np
```

```python
#read files
df_standard = pd.read_csv("Tabela_de_Agentes_Standardizados.csv",sep=";")
df_transport = pd.read_csv("Transport_Document_Data.csv",sep=";")
```

```python
with open("probabilities.json","r") as file:
    prob_dict = json.loads(file.read())

names = list({i for i in df_transport["name"]})
name_split = np.array_split(names,5)

for i in name_split[4]:
    name1 = i.lower()
    name1 = name1.strip()
    min_distance = min([distance(name1,v.strip()) for v in [l.lower() for l in df_standard["name"].values]])
    if i not in prob_dict.keys():
        prob_dict[i] = {"levenshtein":list(),"close_matches":list()}
        for j in df_standard["name"].values:
            name2 = j.lower()
            name2 = name2.strip()

            if distance(name2,name1) <= min_distance + 1:
                if i in prob_dict.keys():
                    prob_dict[i]["levenshtein"].append(j)
            else:
                continue

        prob_dict[i]["close_matches"].append([k for k in get_close_matches(name1,[l.lower() for l in df_standard["name"].values])])

        with open("probabilities.json","w") as file:
            json.dump(prob_dict,file,indent=6)
    else:
        continue
```
